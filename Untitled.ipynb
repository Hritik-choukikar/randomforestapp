```python
import pandas as pd
```

```python
df=pd.read_csv('./data/salaries.csv')
```

```python
df
```

```python
df.head()
```

```python
inputs=df.drop('salary_more_then_100k',axis='columns')
```

```python
target=df['salary_more_then_100k']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lab=LabelEncoder()
```

```python
inputs['company_n']=lab.fit_transform(inputs['company'])
inputs['job_n']=lab.fit_transform(inputs['job'])
inputs['degree_n']=lab.fit_transform(inputs['degree'])
```

```python
inputs
```

```python
inputs_n=inputs.drop(['company','job','degree'],axis='columns')
```

```python
inputs_n
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n.values,target)
```

```python
model.predict([[2,2,1]])
```
